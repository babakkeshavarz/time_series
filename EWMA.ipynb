```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airline_passengers.csv', index_col = 'Month' , parse_dates = True)
```

```python
df.plot()
```

```python
alpha = 0.2
```

```python
df['EWM'] = df['Passengers'].ewm(alpha = alpha , adjust = False).mean()
```

```python
df.plot()
```
